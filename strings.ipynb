```python
msg = "Hello"

print(msg[0])       # prints H
print(msg[4])       # prints o
print(msg[-0])      # prints H, -0 is same as 0
print(msg[-1])      # prints o
print(msg[-2])      # prints l
print(msg[-5])      # prints H
print(msg[-6])      # Index error: string index out of range


```

```python
# A substring can be sliced out of a string

# 1 s[start : end] -extract from start to end -1
print(msg[0:5])

# 2 s[start :] - extract from start to end
print(msg[1:])

# 3 s[: end] - extract from start to end -1
print(msg[0:100])

# 4 s[-start :] - extract from - start(included) to end
print(msg[-5:])

# 5 s[:-end] - extract from begining to -end-1
print(msg[:-2]) # being -2, expression stopped at -3

```

```python
# STRING PROPERTIES

# 1. Python strings are immutable - they can not be changed

message="Hello"
message[0]="M" # Rejected attempt to mutate (change) string

message = "Hii.."
print(message[0])
```

```python
# strings can be concatenated by using +

name="Arjun"
sur_name="Diwedi"

print(name + sur_name)

# strings can be replicated during printing

print("-" * 50)     #prints ---------

#whether on estring is part of another can be found out using (in)

print("e" in "Hello")       # prints True
print("z" in "Hello")       # prints False
print("lo" in "Hello")      # prints True
```

```python
# BUILT IN FUNCTION # ASCCI TABLE

# 1. len(): returns the length of the string, which is the number of characters in the string.

my_string = "Hello, World!"
print(len(my_string))

message = "Racer car 0.1"
print(len(message))

# 2. max(): Returns the character with the highest unicode value ( Referring ASCCI Table) int the string.

my_string = "hello, World!"
print(max(my_string))

passcode = "Unicornx; @ ?!"
print(max(passcode))

# 3. min(): Returns the character with the lowest unicode value ( Referring ASCCI Table) int the string.

my_string = "Hello World!"  # here output would be  space i.e. " " as it has minimum ASCCI value
print(min(my_string))

passcode = "Unicornx; @ ?!" # here output would be  space i.e. " " as it has minimum ASCCI value
print(min(passcode))

```
