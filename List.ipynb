```python
# Aliasing

lst1 = [10,20,30,40,50,60]
lst2 = lst1

index1 = id(lst1)
index2 = id(lst2)



if index1 == index2:
    print("Both the list have same memory address")
else:
    print("Both the list have different memory address")

print(lst1)
print(lst2)

lst1[0]=100

print(lst1)
print(lst2)
```

```python
# 5 POP INDEX 
lst=[1,2,3,4]
print(lst)

tress=lst.pop(1)    # here number 2 from the position 1 from the original list has been popped out
print(tress)    # number 2 will be popped out and printed
print(lst)  # list will be printed without number 2

tress=lst.pop() # as no position numnebr mentioned in the bracket so by default it will pop out last number in the list.
print(tress)
print(lst)

tress=lst.pop(4)    # here 4th position is out of range of original list so occuring error
print(tress)
print(lst)
```
