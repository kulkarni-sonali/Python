```python
# Aliasing

lst1 = [10,20,30,40,50,60]
lst2 = lst1

index1 = id(lst1)
index2 = id(lst2)



if index1 == index2:
    print("Both the list have same memory address")
else:
    print("Both the list have different memory address")

print(lst1)
print(lst2)

lst1[0]=100

print(lst1)
print(lst2)
```
